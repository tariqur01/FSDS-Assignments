# Assignment 1
##### Problem: Write a program that takes the current temperature as input and prints:

```python
current_temp = float(input("Enter current Temperature: "))
if current_temp<10:
    print("It's too cold.")
elif current_temp>=10 and current_temp<=25:
    print("It's cool outside.")
else:
    print("It's hot outside.")


```

# Assignment 2
##### Problem: ATM Withdrawal System

```python
account_balance = float(input("Enter account balance: "))
amount_to_withdraw = float(input("Enter the amount to withdraw: "))

if account_balance >= amount_to_withdraw:
    print("Transaction Successful.")
else:
    print("Insufficient Balance.")
```

# Assignment 3
##### Problem: Student Grade Calculator

```python
mark_obtained = float(input("Enter marks obtained: "))
if mark_obtained<60:
    print("Fail")
elif 60 <= mark_obtained <= 69:
    print("C")
elif 70 <= mark_obtained <= 79:
    print("B")
elif 80 <= mark_obtained <= 89:
    print("A")
elif 90 <= mark_obtained <= 100:
    print("A+")
else :
    print("Invalid Marks")
    
```

# Assignment 4

##### Problem: Bus Ticket Discount

```python
age = int(input("Enter Age: "))
if age<5:
    print("Free")
elif 5 <= age <= 18:
    print("50% Discount") 
elif 19 <= age <= 59:
    print("Full Price")
else :
    print("30% Discount")
```

# Assignment 5
##### Problem: Restaurant Bill with Discount

```python
bill_amount = float(input("Enter bill amount: "))
if bill_amount>1000:
    print(f"10% discount applied. Final bill is: {bill_amount*0.9}")
else:
    print(f"No discount. Final bill is: {bill_amount}")
```

# Assignment 7
##### Problem: Traffic Light System

```python
color = input("Enter color: ")
if color == 'red':
    print('Stop')
elif color == 'yellow':
    print('Ready to go')
elif color == 'green':
    print('Go')
else:
    print('Invalid color')
```

# Assignment 8
##### Problem: Leap Year Checker

```python
year = int(input("Enter year: "))
if year%400 == 0:
    print("Leap Year")
elif year%4 == 0 and not year%100 == 0:
    print("Leap Year")
else:
    print("Not a Leap Year")
```

# Assignment 10
##### Problem: Electricity Bill Calculator

```python
units_consumed = float(input("Enter no of units consumed: "))
if units_consumed<=100:
    print(f"Your bill is 5 BDT per unit. Final bill is {units_consumed*5} BDT")
elif 101<=units_consumed<=200:
    print(f"Your bill is 7 BDT per unit. Final bill is {units_consumed*7} BDT")
else:
    print(f"Your bill is 10 BDT per unit. Final bill is {units_consumed*10} BDT")

```
