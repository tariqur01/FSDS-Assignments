2. Count Word Frequency in Customer Reviews
üíº Real-world use: Analyzing reviews or comments in text analytics.
Task:
Given a review text, count how many times each word appears (ignore case).
Example:
Input: "This product is good and this service is excellent"
Output: {'this': 2, 'product': 1, 'is': 2, 'good': 1, 'and': 1, 'service': 1, 'excellent': 1}

```python
import string

review = input("Enter review: ")
print("Input: ", review)

review_split = [word.strip(string.punctuation) for word in review.lower().split()]
word_count = {}
for word in review_split:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1
print("Output: ", word_count)
```
