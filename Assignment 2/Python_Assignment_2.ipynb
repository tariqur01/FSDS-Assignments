2. Count Word Frequency in Customer Reviews
ðŸ’¼ Real-world use: Analyzing reviews or comments in text analytics.
Task:
Given a review text, count how many times each word appears (ignore case).
Example:
Input: "This product is good and this service is excellent"
Output: {'this': 2, 'product': 1, 'is': 2, 'good': 1, 'and': 1, 'service': 1, 'excellent': 1}

```python
import string

review = input("Enter review: ")
print("Input: ", review)

review_split = [word.strip(string.punctuation) for word in review.lower().split()]
word_count = {}
for word in review_split:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1
print("Output: ", word_count)
```

3. Password Strength Checker 
ðŸ’¼ Real-world use: Used in user authentication systems. 
Task: 
 Check if a password is strong â€” must have at least one uppercase, one lowercase, one digit, and one 
special character. 
Example: 
Input: "Hello@123" 
Output: Strong Password 

```python
import string
password = input("Enter Password: ")
print("Input: ", password)
has_upper = any(c.isupper() for c in password)
has_lower = any(c.islower() for c in password)
has_digit = any(c.isdigit() for c in password)
has_special = any(c in string.punctuation for c in password)

if all([has_upper,has_lower,has_digit,has_special]):
    print("Output: ","Strong Password")
else:
    print("Output: ", "Weak Password")
```

4. Username Generator 
ðŸ’¼ Real-world use: Automatically create usernames for users. 
Task: 
 Take userâ€™s full name and generate a short username. 
 Rules: first 3 letters of first name + last 3 letters of last name + random number (use loops only). 
Example: 
Input: "Mamun Malitha" 
Output: mamitha57

```python
import random
name_raw = input("Enter Name: ")
print("Input: ", name_raw)
name_split = name_raw.split()
username = []
invalid_name = False
if len(name_split)>2:
    print("Name is too long.")
elif len(name_split)<2:
    print("Name is too short.")
elif len(name_split)==2:
    for name in name_split:
        if len(name)<3:
            print("Error: Firstname or Lastname is too short.")
            invalid_name = True
            break
        elif len(name)>3 and name==name_split[0]:
            username.append(name[:3])
        elif len(name)>3 and name==name_split[1]:
            username.append(name[-3:])
    username.append(str(random.randint(10,100)))

print('Username: ', ''.join(username)) if not invalid_name else print("Error: Invalid Name. Username cannot be generated.")

```

6. Remove Duplicates from Contact List 
ðŸ’¼ Real-world use: Data cleaning in CRM or user databases. 
Task: 
 Given a list of names (comma separated), remove duplicates. 
Example: 
Input: "Rahim, Karim, Rahim, Sultana" 
Output: "Rahim, Karim, Sultana"

```python
names = input("Enter the names (Comma Seperated): ")
print("Input: ", names)
names_list = names.split(',')
duplicates_removed = []

for name in names_list:
    if name in duplicates_removed:
        pass
    elif name not in duplicates_removed:
        duplicates_removed.append(name)
print("Output: ", ','.join(duplicates_removed))
```

7. Detect Suspicious Words in Text (Moderation) 
ðŸ’¼ Real-world use: Used in chat moderation or email filtering.

Example:

Output: Warning: Suspicious content detected

```python
msg = input("Enter you message: ")
print("Input: ", msg)
banned_words = ['spam','scam','fake']

for s in banned_words:
    if s in msg.lower():
        print("Warning: Suspicious content detected")
        break
else:
    print("Content is OK")
```

8. Invoice Code Formatter 
ðŸ’¼ Real-world use: Formatting IDs or codes in finance or ERP systems. 
Task: 
 Given an invoice number like 23, generate code in the format INV00023. 
Example: 
Input: 23 
Output: INV00023

```python
number = input("Input: ")
total_digits = 5
padding = []
valid_input = True
if not (number.isdigit() and 0<=int(number)<=int('9'*total_digits)):
    valid_input = False
    
if valid_input:
    sanitized_number = str(int(number))
    print("Input: ", sanitized_number)
    for x in range(total_digits-len(sanitized_number)):
        padding.append('0')
    invoice = ["INV",''.join(padding), sanitized_number]
    print("Output: ", ''.join(invoice))
else:
    print(f"Output: {number} is not a valid number.")
```

9. Count Hashtags and Mentions (Social Media Analytics) 
ðŸ’¼ Real-world use: Used in analyzing social media posts. 
Task: 
 Count how many hashtags (#) and mentions (@) appear in a post. 
Example: 
Input: "Check out #AIWorkshop and follow @inceptionbd" 
Output: Hashtags = 1, Mentions = 1

```python
msg = input("Enter the Message: ")
print("Input: ", msg)
words = msg.split()
hashtags = 0
mentions = 0
for word in words:
    if word.startswith('#') and len(word)>1:
        hashtags += 1
    elif word.startswith('@') and len(word)>1:
        mentions += 1
print("Output: ",f"Hashtags = {hashtags}, Mentions = {mentions}")
```

10. Word Reversal for Data Encryption 
ðŸ’¼ Real-world use: Simple text obfuscation or playful encoding. 
Task: 
 Reverse each word in a sentence while keeping word order the same. 
Example: 
Input: "AI is fun" 
Output: "IA si nuf"

```python
text = input("Enter the Text: ")
words = text.split()
reversed = []
for word in words:
    word = word[::-1]
    reversed.append(word)
print("Output: ", ' '.join(reversed))
```
